```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataku.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
missing_value_count = df.isnull().sum()
```

```python
missing_value_count[0:10]
```

```python
total_cells = np.product(df.shape)
total_missing = missing_value_count.sum()

(total_missing/total_cells) * 100
```

```python
# remove all the rows that contain a missing value
df.dropna()
```

```python
# remove all columns with at least one missing value
columns_with_na_dropped = df.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# just how much data did we lose ?
print("colums in original dataset: %d \n" % df.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```

```python
#Filling in missing values automatically
subset_dataku = df.loc[:, 'EPA':'Season'].head()
subset_dataku
```

```python
#replace all NA's with 0
subset_dataku.fillna(0)
```

```python
# replace all NA's the value that comes directly after it in the same columns
# then replace all the reamining na's with 0
subset_dataku.fillna(method = 'bfill', axis=0).fillna(0)
```
